```python
import json
import pandas as paco # See what I did there
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [10, 6]

# Set up with a higher resolution screen for clarity
%config InlineBackend.figure_format = 'retina'

from datetime import datetime

with open("data-1.json", "r") as f, open("characters.json", "r") as c:
    art_data: dict[str, str | list] = json.load(f)["artwork_data"]
    character_data: dict[str] = json.load(c)['characters']
```

```python
# Render a bar chart that returns the length of each drawings in chronological order
chrono_art = dict()
art_length = []

start_year = 2009
current_year = datetime.now().year

# Get a list of years from 2009 to the current year
y_list = [year for year in range(start_year, current_year+1)]

# Labels for plotting the data
y_lb = "Year"
a_lb = "Art Drawn"

# Iterate for each number of drawings in the y_list variable
for y in y_list:
    y_filter = ['year', str(y)]
    art_num = [d for d in art_data if d['year'] in y_filter]

    art_length.append(len(art_num))

chrono_art.update({y_lb: y_list, a_lb: art_length})

# Plotting the data
df = paco.DataFrame(chrono_art)
df.plot(kind="barh", x=y_lb, y=a_lb,
        title="Total No. of Artworks Drawn In Chronological Order")
```
